# Introduction Reading Files
* How to read a file line by line.
* How to parse text as you read it.

```python
# In the code below, we open a file for reading.  The second argument to the
# open() function is 'r'. This let's python know we want to open the file for
# reading.  Other modes are 'w', to create a new file and write to it, and
# 'a', if we want to open a file and append to the end of it.  I will discuss
# these in lecture.

myFile = open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r')
for line in myFile:
    print(line)
myFile.close()
```

```python
# Why the spaces between each line when we run the cell above?  Well, let's
# use the display function to display a line, without rendering the white
# space. Note the new line escape sequence, '\n' at the end of the line.
display(line)
```

```python
# We can remove white space from a string wit hthe .strip() method.
myFile = open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r')
for line in myFile:
    print(line.strip())
myFile.close()
```

```python
# We can also plit the string by the comma. This will allows us to manipulate
# each field.
myFile = open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r')
for line in myFile:
    lineStrippedSplit = line.strip().split(',')
    print(lineStrippedSplit)
myFile.close()
```

```python
# Now that we have learned how to split up lines, we can proccess the data
# based on the fields:
myFile = open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r')
for line in myFile:
    rank, year, name, count = line.strip().split(',')
    if name == 'William':
        print(count)
myFile.close()
```

```python
# Now that we have learned how to split up lines, we can proccess the data
# based on the fields:
myFile = open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r')
sumCount = 0
for line in myFile:
    rank, year, name, count = line.strip().split(',')
    if name == 'William':
        sumCount += int(count)
myFile.close()
print(sumCount)
```

# Opening A File The "Pythonic" Way

```python
# So, you can open files like we have seen, but it is not "pythonic"
# The following is method is considered better style - we use the with
# keyword
with open(r'C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\baby_name_path\Most_Popular_Baby_Names__1980-2013.csv', 'r') as myFile:
    sumCount = 0
    for line in myFile:
        rank, year, name, count = line.strip().split(',')
        if name == 'William':
            sumCount += int(count)

print(sumCount)
```

# Now, let's look at reading some other files, for practice.
## Melanoma data from data.gov



```python
with open('United States Cancer Statistics, 1999-2011 Incidence.txt', 'r') as myFile:
    count = 0
    for line in myFile:
        print(line.strip())
        count += 1
        if count == 10:
            break
```

```python
display(line)
```

```python
# Notice how there are more columns in the header than in the data. Maybe
# strip() is accidentally remove an empty field from th data.
with open('United States Cancer Statistics, 1999-2011 Incidence.txt',
              'r') as myFile:
    count = 0
    for line in myFile:
        lineFields = line.rstrip().split('\t')
        print(lineFields)
        count += 1
        if count == 10:
            break
```

## Population data from http://seer.cancer.gov/popdata/

This is not a very obvious format! But we can find a description here:
Data description: http://seer.cancer.gov/popdata/popdic.html

```python
with open('us.state.1981_2013.19ages.race.adjusted.txt', 'r') as myFile:
    count = 0
    for line in myFile:
        print(line.strip())
        count += 1
        if count == 10:
            break
```

```python
with open('us.state.1981_2013.19ages.race.adjusted.txt', 'r') as myFile:
    count = 0
    for line in myFile:
        count += 1
        lineStripped = line.strip()
        year = lineStripped[0:4]
        state = lineStripped[4:6]
        population = lineStripped[19:27]
        print(year, state, population)
        if count == 10:
            break
```
