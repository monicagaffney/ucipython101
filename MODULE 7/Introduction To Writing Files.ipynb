## Writing out a csv

```python
lineList = [['CA1432', '123.45'] , ['AZ5287', '938.54'],
            ['FL567', '4125.32']]
with open(r"C:\Users\MGaffney\Documents\LOCALAPPS\PYTHON UCI\MODULE 7\test.csv", "w") as outFile:
    for items in lineList:
        line = ','.join(items) + '\n'
        outFile.write(line)
```

# Reading from one file and writing to another with minimal memory usage

```python
fileToRead = "Most_Popular_Baby_Boy_Names__1980-2013.csv"
fileToWrite = 'Wiliam_Rank.csv'
with open(fileToRead, 'r') as readFile, open(fileToWrite, 'w') as writeFile:
    for line in readFile:
        rank, year, name, count = line.split(',')
        if name == 'William':
            lineToWrite = ','.join([rank, year]) + '\n'
            writeFile.write(lineToWrite)
```
