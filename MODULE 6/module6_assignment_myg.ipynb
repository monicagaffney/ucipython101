Create a function called the_next_15_days. The function will accept one  argument which is a date string in the form of %Y%m%d. Some examples of date strings in this format are ‘20200101’, ‘20210204’, and ‘19831201’. The function will convert this string into a datetime object and then return a list of 15 datetime objects that start on the date that was passed to the function and extend 15 days into the future.  Remember to test your function with different date strings.
Hints

●	You can use list comprehension, although you do not have to.


```python
def the_next_15_days(date):
    d = dt.datetime.strptime(date, "%Y%m%d")
    days = []
    for x in range(0, 15):
        days.append(d + dt.timedelta(days=x))
    return days
```
