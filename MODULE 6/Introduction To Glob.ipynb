# Introduction To The Glob Module
The glob library allows us to select files on the file system based on a pattern in the file path.

    
https://docs.python.org/3/library/glob.html

```python
import glob
import os
```

## Glob Example
* Example: We have several log files with purchase information in a directory.  The log files are named: logdata_<date>.txt, where <date> is a date.



### First, let's get the filepaths to all of the 'logdata' files

```python
data_path = os.path.join('/', 'Users', 'williamhenry', 'Documents', 'Data', 'logdata_*.txt')
print(data_path)

filepaths = glob.glob(data_path)
print(filepaths)
num_files = len(filepaths)
print(num_files)
```

### Now, let's get all of the logdata files from May

```python
data_path = os.path.join('/', 'Users', 'williamhenry', 'Documents', 'Data', 'logdata_*-05-*.txt')

filepaths = glob.glob(data_path)
print(filepaths)
num_files = len(filepaths)
print(num_files)
```
